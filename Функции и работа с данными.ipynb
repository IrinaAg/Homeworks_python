Напишите функцию, которая классифицирует фильмы из материалов занятия по правилам:

    оценка 2 и ниже — низкий рейтинг;
    оценка 4 и ниже — средний рейтинг;
    оценка 4.5 и 5 — высокий рейтинг.

Результат классификации запишите в столбец class.

```python
import pandas as pd
```

```python
data = pd.read_csv('ratings.csv', sep=',')
```

```python
def raiting_class(row):
    if row.rating >= 4.5:
        return 'высокий рейтинг'
    elif 2 < row.rating < 4.5:
        return 'средний рейтинг'
    elif row.rating <= 2:
        return 'низкий рейтинг'
        
data['class'] = data.apply(raiting_class, axis=1)
data.loc[150:155]
```

```python
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
towns = []

towns = []
for values in geo_data.values():
    for i in values:
        towns.append(i)
towns
```

```python
row = 'вк петербург'
data = row.split(' ')
def check(t):
    for i in t:
        if i in towns:
            for items in geo_data.items():
                if i in items[1]:
                    print(items[0])
check(data)
```

```python
df = pd.read_csv('keywords.csv')
df.head()
```

```python
def region_in_keyword(row):
    data = row['keyword'].split(' ')
    i = 0
    for word in data:
        if word in towns:
            for items in geo_data.items():
                if word in items[1]:
                    i += 1
                    return items[0]
    if i == 0:
        return 'undefined'       
```

```python
df['region'] = df.apply(region_in_keyword, axis=1)
df.head()
```

```python
years = list(range(1950, 2011))
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
joined = ratings.merge(movies, on='movieId', how='left')
```

```python
def record_of_year(x):
    for year in years:
        if str(year) in x:
            return str(year)
    return '1900'
```

```python
joined['year'] = joined['title'].apply(record_of_year)
joined.head()
```

```python
joined.groupby('year').mean()[['rating']].sort_values('rating', ascending=False).head(10)
```
